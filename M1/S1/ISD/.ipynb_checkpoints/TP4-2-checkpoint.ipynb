```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Charger votre dataset
df = pd.read_csv('unpopular_songs_full.csv')

# Supprimer la classe 'popularity' si elle est prÃ©sente
if 'popularity' in df.columns:
    df = df.drop('popularity', axis=1)

# Normaliser les donnÃ©es si nÃ©cessaire
# Utilisez uniquement des colonnes numÃ©riques pour le clustering
X = df.select_dtypes(include=[np.number])

# Initialisez une liste pour stocker les valeurs de la silhouette
silhouette_scores = []

# Essayez diffÃ©rents nombres de clusters
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X)
    cluster_labels = kmeans.predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Tracer la courbe de la silhouette
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Nombre de clusters')
plt.ylabel('Score de silhouette')
plt.title('MÃ©thode de la silhouette pour dÃ©terminer le nombre de clusters')
plt.show()
```

```python
import numpy as np

def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2)**2))

def centroid(cluster):
    return np.mean(cluster, axis=0)

def cluster_dispersion(cluster, centroid):
    return np.max([euclidean_distance(point, centroid) for point in cluster])

def davies_bouldin_score(clusters):
    k = len(clusters)
    centroids = [centroid(cluster) for cluster in clusters]
    cluster_distances = np.zeros((k, k))

    for i in range(k):
        for j in range(i+1, k):
            intra_cluster_distance = euclidean_distance(centroids[i], centroids[j])
            average_dispersion = (cluster_dispersion(clusters[i], centroids[i]) + cluster_dispersion(clusters[j], centroids[j])) / 2
            cluster_distances[i, j] = cluster_distances[j, i] = average_dispersion / intra_cluster_distance

    max_cluster_distances = np.max(cluster_distances, axis=1)
    return np.mean(max_cluster_distances)

# Exemple d'utilisation :
# clusters est une liste de clusters, oÃ¹ chaque cluster est reprÃ©sentÃ© par une liste de points
# davies_bouldin_score_value = davies_bouldin_score(clusters)
# print("Davies-Bouldin Score:", davies_bouldin_score_value)
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score

def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2)**2))

def centroid(cluster):
    return np.mean(cluster, axis=0)

def cluster_dispersion(cluster, centroid):
    return np.max([euclidean_distance(point, centroid) for point in cluster])

def davies_bouldin_score_from_kmeans(X, num_clusters):
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X)
    
    labels = kmeans.labels_
    clusters = [X[labels == i] for i in range(num_clusters)]

    k = len(clusters)
    centroids = [centroid(cluster) for cluster in clusters]
    cluster_distances = np.zeros((k, k))

    for i in range(k):
        for j in range(i+1, k):
            intra_cluster_distance = euclidean_distance(centroids[i], centroids[j])
            average_dispersion = (cluster_dispersion(clusters[i], centroids[i]) + cluster_dispersion(clusters[j], centroids[j])) / 2
            cluster_distances[i, j] = cluster_distances[j, i] = average_dispersion / intra_cluster_distance

    max_cluster_distances = np.max(cluster_distances, axis=1)
    return np.mean(max_cluster_distances)

# Exemple d'utilisation :
# Remplacez X par votre ensemble de donnÃ©es.

num_clusters = 3  # Remplacez par le nombre de clusters que vous voulez.
db_score = davies_bouldin_score_from_kmeans(X, num_clusters)
print("Davies-Bouldin Score:", db_score)
```
